```python
import shutil

# Define the paths
dst_path_train = r"/kaggle/working/Chula-ParasiteEgg-11/images/train"
dst_path_val = r"/kaggle/working/Chula-ParasiteEgg-11/images/validation"

# Function to delete a folder if it exists
def delete_folder(path):
    if os.path.exists(path):
        shutil.rmtree(path)
        print(f"Deleted {path}")
    else:
        print(f"{path} does not exist.")

# Delete the 'train' and 'validation' folders
delete_folder(dst_path_train)
delete_folder(dst_path_val)
```

```python
import random
import shutil
import os

src_path = r"/kaggle/input/chula-parasite-dataset/Chula-ParasiteEgg-11/Chula-ParasiteEgg-11/Chula-ParasiteEgg-11/data"
dst_path_train = r"/kaggle/working/Chula-ParasiteEgg-11/images/train"
dst_path_val = r"/kaggle/working/Chula-ParasiteEgg-11/images/validation"

# Create the destination directories if they don't exist
if not os.path.exists(dst_path_train):
    os.makedirs(dst_path_train)
if not os.path.exists(dst_path_val):
    os.makedirs(dst_path_val)

# Get a list of all images in the source directory
image_paths = [os.path.join(src_path, f) for f in os.listdir(src_path) if os.path.isfile(os.path.join(src_path, f))]

# Randomly shuffle the image paths
random.shuffle(image_paths)

# Split the image paths into train and validation sets
train_image_paths = image_paths[:int(0.8 * len(image_paths))]
val_image_paths = image_paths[int(0.8 * len(image_paths)):]

# Copy the train images to the train destination directory
for image_path in train_image_paths:
    shutil.copy(image_path, dst_path_train)

# Copy the validation images to the validation destination directory
for image_path in val_image_paths:
    shutil.copy(image_path, dst_path_val)

print('Copied')
```

```python
import json
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
file = open('/kaggle/input/chula-parasite-dataset/Chula-ParasiteEgg-11/Chula-ParasiteEgg-11/Chula-ParasiteEgg-11/labels.json')
data = json.load(file)
```

```python
image_df = pd.DataFrame.from_dict(pd.json_normalize(data['images']), orient='columns')
image_df
```

```python
annotation_df = pd.DataFrame.from_dict(pd.json_normalize(data['annotations']), orient='columns')
annotation_df
```

```python
duplicate_values = annotation_df['image_id'].duplicated()
duplicate_values
```

```python
merged_df = pd.merge(image_df, annotation_df, left_on='id', right_on='image_id', how='inner')

# Drop the extra 'image_id' column as it's now redundant
merged_df.drop(columns=['image_id'], inplace=True)
merged_df
```

```python
# Calculate bbox_yolo values
merged_df['bbox_yolo'] = merged_df.apply(lambda row: [
    ((row['bbox'][0] + row['bbox'][2] / 2) / row['width']),
    ((row['bbox'][1] + row['bbox'][3] / 2) / row['height']),
    (row['bbox'][2] / row['width']),
    (row['bbox'][3] / row['height'])
], axis=1)

# Display the DataFrame with the new bbox_yolo field
merged_df
```

```python
import pandas as pd
import os

output_dir = '/kaggle/working/Chula-ParasiteEgg-11/labels/train'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Iterate through the rows and create text files
for index, row in merged_df.iterrows():
    category_id = row['category_id']
    bbox = row['bbox_yolo']
    filename = os.path.splitext(row['file_name'])[0]  # Remove the '.jpg' extension
    
    # Create the text content
    text_content = f"{category_id} {' '.join(map(str, bbox))}"
    
    # Write the text content to a file
    output_filename = os.path.join(output_dir, f"{filename}.txt")
    with open(output_filename, 'w') as text_file:
        text_file.write(text_content)
```
